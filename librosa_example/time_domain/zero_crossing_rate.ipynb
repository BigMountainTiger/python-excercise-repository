```python
# https://en.wikipedia.org/wiki/Zero-crossing_rate

import os
import librosa
import librosa.display
import IPython.display as idp
import matplotlib.pyplot as plt
import numpy as np
```

```python
wav = '../audio/my-song.wav'
waveform, sr = librosa.load(wav)
```

```python
idp.Audio(wav)
```

```python
FRAME_SIZE = 1024
HOP_LENGTH = 512

my_song_zcr = librosa.feature.zero_crossing_rate(waveform, FRAME_SIZE, HOP_LENGTH)[0]
print(f'Total {my_song_zcr.size} points')
```

```python
frames = range(0, my_song_zcr.size)
t = librosa.frames_to_time(frames, hop_length = HOP_LENGTH)

plt.figure(figsize=(15,5))
librosa.display.waveplot(waveform, alpha=0.5)
plt.plot(t, my_song_zcr, color='red')
plt.title('My song')
plt.ylim((-1, 1))
plt.show()
```
